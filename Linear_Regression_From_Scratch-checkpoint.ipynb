```python
X = df.drop(columns=["TOTCHG"])
```

```python
X = X.values
```

```python
X = X.astype(float)
```

```python
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)

X_std[X_std == 0] = 1

X = (X - X_mean) / X_std
```

```python
X.shape
```

```python
y.shape
```

```python
n = X.shape[1]
theta = np.zeros((n, 1))
```

```python
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X @ theta
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost
```

```python
def gradient_descent(X, y, theta, alpha, epochs):
    m = len(y)
    cost_history = []
    metrics_history = []

    for epoch in range(epochs):
        predictions = X @ theta
        errors = predictions - y

        gradients = (1 / m) * (X.T @ errors)
        theta = theta - alpha * gradients

        cost = compute_cost(X, y, theta)
        cost_history.append(cost)

        rmse = np.sqrt(np.mean(errors ** 2))
        r2 = 1 - (np.sum(errors ** 2) / np.sum((y - np.mean(y)) ** 2))

        metrics_history.append((rmse, r2))

        print(
            f"Epoch {epoch+1}: "
            f"Cost={cost:.4f}, RMSE={rmse:.4f}, R2={r2:.4f}"
        )

    return theta, cost_history, metrics_history
```

```python
alpha = 0.01
epochs = 100

theta, cost_history, metrics_history = gradient_descent(
    X, y, theta, alpha, epochs
)
```

```python
plt.plot(range(epochs), cost_history)
plt.xlabel("Epochs")
plt.ylabel("Cost (MSE)")
plt.title("Cost Function vs Epochs")
plt.show()
```
